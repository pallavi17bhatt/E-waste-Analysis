# Frequency of E-Waste Disposal
From this graph we can easily see that majority of people are not disposing off their e-waste for longer than an year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file
file_path = '/content/sample_data/E-Waste Management Survey_response_latest.csv'


data = pd.read_csv(file_path)

# Analyze the frequency of E-waste disposal
e_waste_disposal_frequency = data['How often do you dispose of electronic devices (e.g., phones, computers, appliances)?'].value_counts()

# Plotting the data
plt.figure(figsize=(10, 6))  # Set the figure size
e_waste_disposal_frequency.plot(kind='bar', color='skyblue', alpha=0.7)  # Create a bar plot
plt.title('Frequency of E-waste Disposal')  # Title of the plot
plt.xlabel('Frequency')  # Label for the X-axis
plt.ylabel('Number of Responses')  # Label for the Y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for the y-axis
plt.show()  # Display the plot
```

# Mode Values: Most Common Response by Questions


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from a CSV file
file_path = '/content/sample_data/E-Waste Management Survey_response_latest.csv'  # Ensure to use the correct path to your CSV file
data = pd.read_csv(file_path)

# Calculate the mode for each column in the dataset
mode_data = data.mode().iloc[0]

# Prepare data for plotting
mode_df = pd.DataFrame(mode_data)
mode_df.reset_index(inplace=True)
mode_df.columns = ['Survey Question', 'Most Common Response']

# Plotting using matplotlib and seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Most Common Response', y='Survey Question', data=mode_df, palette='viridis')

plt.title('Mode of Survey Responses')
plt.xlabel('Most Common Response')
plt.ylabel('Survey Question')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjusts plot parameters to give a pleasing layout
plt.show()
```

 # Awareness of Environmental Impact
 Nearly 75% of the people of do not care or are unware about the environmental Imapcts of E waste.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the survey data
file_path = '/content/sample_data/E-Waste Management Survey_response_latest.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Count the frequency of each response category for the awareness question
awareness_counts = data['How aware are you of the environmental impact of improper e-waste disposal?'].value_counts()

# Create a pie chart for the awareness about e-waste disposal
plt.figure(figsize=(8, 8))
plt.pie(awareness_counts, labels=awareness_counts.index, autopct='%1.1f%%', colors=sns.color_palette('coolwarm', n_colors=len(awareness_counts)))
plt.title('Awareness of the Environmental Impact of Improper E-Waste Disposal')
plt.show()
```

# Barriers to Proper Disposal of E-Waste Disposal

1. About 40% people are not aware or educated about the disposal methods.
2. Nearly 56% people are unable to do it due to lack of proper processes and disposal methods.  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/sample_data/E-Waste Management Survey_response_latest.csv')
disposal_method_counts=data['What is the primary barrier to proper e-waste disposal for you?'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(disposal_method_counts, labels=disposal_method_counts.index, autopct='%1.1f%%', colors=sns.color_palette('coolwarm', n_colors=len(disposal_method_counts)))
plt.title('Barriers to Proper Disposal of E-Waste Disposal')
plt.show()
```

# Satisfaction with E-Waste Management Services
1. About 45 % people does not have any opinion of the services, methods of E-waste disposal.
2. 37% people are not at all satisfied with the disposal services in their area.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/sample_data/E-Waste Management Survey_response_latest.csv')
satisfaction_counts=data['How satisfied are you with the e-waste management services in your area?'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.1f%%', colors=sns.color_palette('coolwarm', n_colors=len(satisfaction_counts)))
plt.title('Satisfaction with E-Waste Management Services')
plt.show()
```
